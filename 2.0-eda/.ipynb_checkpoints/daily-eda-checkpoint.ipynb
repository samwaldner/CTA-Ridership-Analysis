```python
# Code modified from code written by Matthew Garton.

def plot_series(df, cols=None, title='Title', xlab=None, ylab=None, steps=1):
    
    # Set figure size to be (18, 9).
    plt.figure(figsize=(18,9))
    
    # Iterate through each column name.
    for col in cols:
            
        # Generate a line plot of the column name.
        # You only have to specify Y, since our
        # index will be a datetime index.
        plt.plot(df[col])
        
    # Generate title and labels.
    plt.title(title, fontsize=26)
    plt.xlabel(xlab, fontsize=20)
    plt.ylabel(ylab, fontsize=20)
    
    # Enlarge tick marks.
    plt.yticks(fontsize=18)
    plt.xticks(df.index[0::steps], fontsize=18);
```

```python
plot_series(df,['total_rides'], title='Rides',steps=850)
```

# Total Rides

```python
decomp = seasonal_decompose(df['total_rides'])
```

```python
rcParams['figure.figsize'] = (12, 8)
decomp.plot();
```

The trend line is going to be a problem as it looksl more like the Seasonal line should, and doesn't actually show direction until 2020.

Next we are going to create an Autocorrelation and some Partial Autocorrelation charts to help determine appropriate lag length as well as to better detect Seasonality and trend. Here trend seems to be less evident than seasonality.

```python
plot_acf(rides['total_rides'],lags=9);
```

```python
plot_pacf(rides['total_rides'], lags=30,method='ywm');
plot_pacf(rides['total_rides'], lags=90,method='ywm');
plot_pacf(rides['total_rides'], lags=375,method='ywm');
```

We can see clear seasonality with weekly peaks and stronger partial autocorrelation closer to 1 lag and that 7-day lag mark.
