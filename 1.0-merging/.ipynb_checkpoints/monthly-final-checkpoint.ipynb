```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)
```

```python
auto = pd.read_excel('../../capstone-data/monthly/agg_auto.xlsx')
weekly = pd.read_csv('../../capstone-data/weekly-merged.csv')
```

# Preop Weekly as Monthly

```python
weekly.head()
```

```python
weekly['date']=pd.to_datetime(weekly['date'])
```

```python
weekly.set_index('date',inplace=True)
weekly.sort_index(inplace=True)
```

```python
weekly.head()
```

```python
weekly = weekly.resample('MS').agg({'bus':'sum', 'rail_boardings':'sum', 'total_rides':'sum', 'prcp':'mean', 'tmax':'mean', 'tmin':'mean',
       'gas_open':'mean', 'gas_close':'mean', 'gas_volume':'mean', '3_mo':'mean', '6_mo':'mean',
        '1_yr':'mean', '2_yr':'mean',
       '3_yr':'mean', '5_yr':'mean', '7_yr':'mean', '10_yr':'mean', '20_yr':'mean', 'C1_School closing':'mean',
       'C2_Workplace closing':'mean', 'C3_Cancel public events':'mean',
       'C4_Restrictions on gatherings':'mean', 'C5_Close public transport':'mean',
       'C6_Stay at home requirements':'mean', 'C7_Restrictions on internal movement':'mean',
       'C8_International travel controls':'mean', 'E1_Income support':'mean',
       'E2_Debt/contract relief':'mean', 'H1_Public information campaigns':'mean',
       'H2_Testing policy':'mean', 'H3_Contact tracing':'mean', 'H6_Facial Coverings':'mean',
       'H7_Vaccination policy':'mean', 'H8_Protection of elderly people':'mean',
       'ConfirmedCases':'sum', 'ConfirmedDeaths':'sum', 'StringencyIndex':'mean',
       'StringencyLegacyIndex':'mean', 'GovernmentResponseIndex':'mean',
       'ContainmentHealthIndex':'mean', 'EconomicSupportIndex':'mean', 'total_daily_doses':'mean',
       'total_daily_cum':'sum', 'first_dose_daily':'mean', 'first_dose_cum':'sum',
       'first_dose_percent_pop':'mean', 'vax_series_completed_daily':'mean',
       'vax_series_cum':'sum', 'vax_series_percent':'mean', 'crash_occurrences':'sum',
       'damage_indicator':'mean','gas_price':'mean'})
```

```python
weekly.head(3)
```

```python
weekly.isna().sum()
```

```python
weekly.index.duplicated().sum()
```

# Prep Auto

```python
auto.head()
```

```python
auto.rename(columns={'Unnamed: 0':'date'},inplace=True)
```

```python
auto = auto[['date','new_light_truck_reg','new_car_reg','new_total_reg']]
```

```python
auto.dtypes
```

```python
auto.set_index('date',inplace=True)
auto.sort_index(inplace=True)
```

```python
auto.head()
```

```python
auto = auto.asfreq('MS')
```

```python
auto.isna().sum()
```

```python
auto.index.duplicated().sum()
```

```python
auto.index
```

## Merge Weekly & Auto

```python
merged = weekly.merge(auto, how='outer', left_index=True, right_index=True)
```
